```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ["FY2017-18", "FY2018-19", "FY2019-20", "FY2020-21", "FY2021-22", "FY2022-23", "FY2023-24\n(till 31st Dec'23)"]
targets = [2500, 3013, 4019, 5500, 6000, 13233, 18000]
achievements = [2071, 3134, 4572, 5554, 8839, 13462, 13296]
growth = [0,51, 46, 21, 59, 52]  # The last value is set to 0 as the chart does not have it

# Bar chart configuration
bar_width = 0.35
x = np.arange(len(years))

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting the bar chart
bars1 = ax1.bar(x - bar_width/2, targets, bar_width, label='Target', color='blue')
bars2 = ax1.bar(x + bar_width/2, achievements, bar_width, label='Achievement', color='orange')

# Adding labels for bars
ax1.set_xlabel('Financial Year')
ax1.set_ylabel('In Crore')
ax1.set_title('Digital Transaction Target vs. Achievement (in crore)')
ax1.set_xticks(x)
ax1.set_xticklabels(years)
ax1.legend(loc='upper left')

# Adding data labels on the bars
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=9, color='blue')

for bar in bars2:
    height = bar.get_height()
    ax1.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=9, color='orange')

# Creating a twin Axes sharing the xaxis
ax2 = ax1.twinx()
ax2.set_ylabel('% Growth')

# Plotting the line chart
line = ax2.plot(x[:-1], growth, label='% Growth', color='brown', marker='o', linestyle='-')

# Adding data labels for the line chart
for i, txt in enumerate(growth):
    ax2.annotate(f'{txt}%', (x[i], growth[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9, color='brown')

ax2.legend(loc='upper right')

# Show plot
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ["Food & beverages", "Pan, tobacco & intoxicants", "Clothing & footwear", 
              "Housing", "Fuel and light", "Miscellaneous"]
FY21 = [54, 5, 4, 7, 3, 30]
FY22 = [35.0, 3, 8, 8.0, 12, 34]  # Example data for FY22
FY23 = [47, 2, 9, 7, 7.5, 25.0]  # Example data for FY23
weights = [45.9, 2.4, 6.5, 10.1, 6.8, 28.3]  # Example weights

# Define colors
colors = ['#ff9999','#66b3ff','#99ff99']

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(12, 7))

# Bar width
bar_width = 0.25
x = np.arange(len(categories))

# Plotting the bar chart
bars1 = ax1.bar(x - bar_width, FY21, bar_width, label='FY21', color=colors[0])
bars2 = ax1.bar(x, FY22, bar_width, label='FY22', color=colors[1])
bars3 = ax1.bar(x + bar_width, FY23, bar_width, label='FY23*', color=colors[2])

# Adding labels for bars
ax1.set_xlabel('Categories', fontsize=12)
ax1.set_ylabel('Contribution (per cent)', fontsize=12)
ax1.set_title("Retail Inflation Driven by 'Food and Beverages' Group", fontsize=14)
ax1.set_xticks(x)
ax1.set_xticklabels(categories, fontsize=10, rotation=45)
ax1.tick_params(axis='y', labelsize=10)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)

# Creating a twin Axes sharing the xaxis
ax2 = ax1.twinx()
ax2.set_ylabel('Weights', fontsize=12)
ax2.tick_params(axis='y', labelsize=10)

# Plotting the line chart for weights
line = ax2.plot(x, weights, label='Weights', color='blue', marker='D', linestyle='-', markersize=8)

# Adding data labels for the line chart
for i, txt in enumerate(weights):
    ax2.annotate(f'{txt}', (x[i], weights[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='blue')

# Add legend at the bottom
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=4, fontsize=10)

# Adjust layout
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data: Month and corresponding Inflation Rate
months = ['Oct-21', 'Nov-21', 'Dec-21', 'Jan-22', 'Feb-22',
          'Mar-22', 'Apr-22', 'May-22', 'Jun-22', 'Jul-22',
          'Aug-22', 'Sep-22', 'Oct-22']
inflation_rates = [4.48, 4.91, 5.66, 6.01, 6.07,
                   6.95, 7.79, 7.04, 7.01, 6.71,
                   7.00, 7.41, 6.77]

# Plotting the line chart with data labels
plt.plot(months, inflation_rates, marker='o', label='Inflation Rate', color='brown', linestyle='-.', linewidth=2)
plt.xlabel('Month')
plt.ylabel('Inflation Rate (%)')
plt.title('All India Year on Year Inflation Rates based on CPI', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()

# Adding data labels
for i in range(len(months)):
    plt.text(months[i], inflation_rates[i], f"{inflation_rates[i]:.2f}%",
             ha='center', va='bottom', fontsize=9, color='black')

# Customize the appearance
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left')
plt.axhline(y=6.5, color='r', linestyle='--', label='Threshold (6.5%)')
plt.fill_between(months, inflation_rates, 6.5, where=(inflation_rates >= 6.5), color='red', alpha=0.2)
plt.annotate('High Inflation', xy=(6, 6.7), xytext=(8, 7.2),
             arrowprops=dict(facecolor='black', shrink=0.05, width=0.5, headwidth=6),
             fontsize=9, color='red')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Commitments Raised', 'Funds Raised', 'Investments Made']
years = ['2018-19', '2019-20', '2020-21', '2021-22', '2022-23']
values = [
    [2.8, 3.7, 8.3, 4.5, 6.4],  # Commitments Raised
    [1.3, 1.9, 2.3, 3.1, 3.7],  # Funds Raised
    [1.1, 1.5, 2.0, 2.8, 3.4]   # Investments Made
]

# Number of bars per category
n_bars = len(years)

# Positions of the bars on the x-axis
bar_width = 0.15
bar_positions = np.arange(len(categories))

# Create figure and axis
fig, ax = plt.subplots()

# Plot bars
for i, year in enumerate(years):
    ax.bar(bar_positions + i * bar_width, [value[i] for value in values], width=bar_width, label=year)

# Set labels
ax.set_xlabel(' ')
ax.set_ylabel('₹ lakh crore')
ax.set_title('Growing Alternative Investment Fund Ecosystem')
ax.set_xticks(bar_positions + bar_width * (n_bars - 1) / 2)
ax.set_xticklabels(categories)
ax.legend(title='Financial Year')

# Note
plt.figtext(0.5, -0.05, 'Note: The above numbers are cumulative net figures at the end of each financial year', ha='center', fontsize=9)

# Display the plot
plt.show()
```
